```python
fill_cuisines = df['Cuisines'].fillna('Unknown',inplace = True)
```

```python
#count the occurences of Cuisines 
cuisine_counts = df['Cuisines'].str.split(',').explode('Cuisines').value_counts()
print(cuisine_counts)
top_3_cuisines = cuisine_counts.head(3)

print("The top 3 cuisines are :")
print (top_3_cuisines)
```

```python
colours = ['yellow','orange','blue']
plt.bar(top_3_cuisines.index, top_3_cuisines.values,color= colours)
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Top 3 Cuisines')
plt.show()
```

```python
# calculate the percentage of restaurants that serve each of the top cuisines.
total_restaurants = len(df)
print(total_restaurants)
percentage = (cuisine_counts/total_restaurants)*100

print("The market share of top 5 cuisine are :")
top_5_pct = percentage.head(5)
print(top_5_pct)
```

```python
plt.bar(top_5_pct.index,top_5_pct.values,color = colours)
plt.xlabel('Cuisines')
plt.ylabel('percentage')
plt.title('Top 5 cuisines %')
plt.show()
```
